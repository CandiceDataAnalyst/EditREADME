#Step 1: Removing duplicate rows ( there could be duplicate rows excluding Row_ID column ).
https://www.w3schools.com/python/ref_set_difference.asp
https://www.digitalocean.com/community/tutorials/pandas-drop-duplicate-rows-drop_duplicates-function

```python
# Drop duplicate rows based on all columns except for Row_ID
df = df.drop_duplicates(subset=df.columns.difference(['Row_ID'])) #df.columns.difference(['Row_ID']) returns a list of all column names in the DataFrame except for Row_ID
```

```python
df.info()
```

#Step 2: Removing rows for which few values are missing.
#https://www.w3resource.com/pandas/dataframe/dataframe-dropna.php

```python
#find the column with the number of missing values.
df.isna().sum()
```

```python
#using dropno ()function to remove missing values
df=df.dropna()
```

```python
#check if missing values were sucessfully dropped 
df.isna().sum()
```

#Step 3: Remove irrelevant values from each column if any. Validation of all values for a column( order date and ship date value must be in correct date format ). For each entry in dataset ship date >= order date
https://www.geeksforgeeks.org/python-pandas-to_datetime/
https://pypi.org/project/pycountry/

```python
import datetime

# remove irrelevant values and validate date format
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='%d/%m/%Y', errors='coerce')
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'], format='%d/%m/%Y', errors='coerce')

# drop rows where ship date is earlier than order date
df = df[df['Ship_Date'] >= df['Order_Date']]
df
```

```python

# Loop through each column in the DataFrame
for col in df.columns:
    print(f'Processing column: {col}')
    
    # Check if the column is a date column
    if col in ['Order_Date', 'Ship_Date']:
        print(f'Validating date format for column: {col}')
        # Validate the date format
        df[col] = pd.to_datetime(df[col],  format='%d/%m/%Y', dayfirst=True, errors='coerce')
        print(f'Checking ship date is greater than or equal to order date for column: {col}')
        # Check if the ship date is greater than or equal to the order date
        df = df[df['Ship_Date'] >= df['Order_Date']]
        
    else:
        # Remove any rows that have an irrelevant value in the column
        if col == 'Country':
            relevant_countries = [country.name.lower() for country in pycountry.countries]
            print(f'Removing irrelevant rows for column: {col}')
            df = df[df[col].str.lower().isin(relevant_countries)]
        else:
            # You can add more elif statements for other columns that need cleaning/validation
            pass

```
