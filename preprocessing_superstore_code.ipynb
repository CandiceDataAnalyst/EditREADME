```python
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
pd.options.display.max_columns = None
```

```python
#read the file
df = pd.read_csv("/Users/candicewu/Desktop/superstore_final_dataset.csv")
df.head()
```

```python
df.info()
```

#### Step 1: drop unnecessary columns </p>
https://www.geeksforgeeks.org/how-to-drop-one-or-multiple-columns-in-pandas-dataframe/

```python
#drop unneceessary column
df=df.drop(['Row_ID'],axis=1)
df
```

```python
df= df.drop_duplicates()
df.info()
```

#### Step 2: Removing rows for which few values are missing</p>
https://www.w3resource.com/pandas/dataframe/dataframe-dropna.php

```python
#find the column with the number of missing values.
df.isna().sum()
```

```python
#using dropno ()function to remove missing values
df=df.dropna()
df.isna().sum()
```

#### Step 3: Remove irrelevant values from each column if any. Validation of all values for a column( order date and ship date value must be in correct date format ) For each entry in dataset ship date >= order date </p>
https://www.geeksforgeeks.org/python-pandas-to_datetime/
https://pypi.org/project/pycountry/

```python
import datetime

# remove irrelevant values and validate date format
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='%d/%m/%Y', errors='coerce')
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'], format='%d/%m/%Y', errors='coerce')

# drop rows where ship date is earlier than order date
df = df[df['Ship_Date'] >= df['Order_Date']]
df
```

```python
df=df.sort_values(by="Order_Date")
df
```

#### Step 4: Export the cleaned dataset as a .csv file: prefer UTF-8 encoding.

```python
df.to_csv('/Users/candicewu/Desktop/cleaned_superstore_dataset.csv', index=False)
```
